# Comparing Clustering Algorithms

### 1. Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import unique
from numpy import where
from sklearn.datasets import make_classification
from matplotlib import pyplot
```

### 2. Loading dataset

```python
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)

# create scatter plot for samples from each class
for class_value in range(2):
    row_ix = where(y == class_value) # get row indexes for samples with this class
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
    
pyplot.show() # show the plot
```

### 3. Models

```python
import warnings
warnings.filterwarnings("ignore")
```

#### 3.1. Affinity propagation clustering

```python
from sklearn.cluster import AffinityPropagation

# dataset
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
```

```python
model = AffinityPropagation(damping=0.9) # define the model
model.fit(X) # fit the model
yhat = model.predict(X) # assign a cluster to each example
clusters = unique(yhat) # retrieve unique clusters
```

```python
# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
    
# show the plot
pyplot.show()
```

#### 3.2. AgglomerativeClustering

```python
from sklearn.cluster import AgglomerativeClustering

# define dataset
X, _ = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=4)
```

```python
model = AgglomerativeClustering(n_clusters=2) # define the model
yhat = model.fit_predict(X) # fit model and predict clusters
clusters = unique(yhat) # retrieve unique clusters
```

```python
# create scatter plot for samples from each cluster
for cluster in clusters:
    row_ix = where(yhat == cluster) # get row indexes for samples with this cluster
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1]) # create scatter of these samples
    
# show the plot
pyplot.show()
```
